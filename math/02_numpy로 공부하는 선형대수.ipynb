### 분배법칙은 성립한다
$ A(B + C) = AB + AC  $

````python
A @ (B + C)
````

````python
A @ B + A @ C
````

````python
A @ (B + C) == A @ B + A @ C
````

````python
A @ C + B @ C
````

````python
(A + B) @ C
````

````python
A @ C + B @ C == (A + B) @ C
````

````python
# 2.4 선형 연립방정식과 역행렬
# 넘파이를 사용한 역행렬 계산
A = np.array([[1,1,0],[0,1,1],[1,1,1,]])
A
````

````python
Ainv = np.linalg.inv(A)
Ainv
````

````python
# 역행렬과 선형 연립방정식의 해
# x = A(역행렬) * b
b = np.array([[2],[2],[3]])
b
````

````python
x = Ainv @ b
x
````

````python
# x벡터를 원래의 연립방정식에 대입하여 상수벡터B와 값이 일치하는지 확인
A @ x - b
````

````python
# lstsq 명령어로도 가능함
x , resid, rank, s = np.linalg.lstsq(A, b)
x
````
