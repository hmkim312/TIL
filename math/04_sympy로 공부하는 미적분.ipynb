```python
# 부호 함수
np.sign(-11), np.sign(0), np.sign(9)
```

```python
# 단위 계단 함수
def heaviside_step(x):
    if isinstance(x, np.ndarray):
        return np.where (x >= 0, 1, 0)
    else :
        return 1.0 if x >= 0 else 0.0
heaviside_step(-1), heaviside_step(0), heaviside_step(2)
```

```python
# 함수의 그래프
def f(x) :
    return x ** 3 - 3 * x ** 2 + x
x = np.linspace(-1,3,9)
y = f(x)
y
```

```python
x = np.linspace(-1,3,9)
y = f(x)
plt.plot(x, y, 'ro-')
plt.xlim(-2, 4)
plt.title("함수 $f(x) = x^3 - 3x^2 + x$의 그래프")
plt.xlabel('x')
plt.ylabel('y')
plt.xticks(np.arange(-1, 4))
plt.yticks(np.arange(-5, 4))
plt.show()
```

```python
x = np.linspace(-1,3,400)
y = f(x)
plt.figure(figsize=(5,3))
plt.plot(x, y, 'r')
plt.xlim(-2, 4)
plt.title("함수 $f(x) = x^3 - 3x^2 + x$의 그래프")
plt.xlabel('x')
plt.ylabel('y')
plt.xticks(np.arange(-1, 4))
plt.yticks(np.arange(-5, 4))
plt.show()
```

```python
# 연습문제 4.1.1
# 부호함수
x = np.linspace(-1,3,100)
y = np.sign(x)
plt.figure(figsize=(5,3))
plt.plot(x, y, 'r')
plt.xlim(-2, 4)
plt.title("부호함수")
plt.xlabel('x')
plt.ylabel('y')
plt.xticks(np.arange(-1, 3))
plt.yticks(np.arange(-2, 3))
plt.show()
```

```python
# 단위 계단 함수
x = np.linspace(-1,3,100)
y = heaviside_step(x)
plt.figure(figsize=(5,3))
plt.plot(x, y, 'r')
plt.xlim(-2, 4)
plt.title("단위 계단함수")
plt.xlabel('x')
plt.ylabel('y')
plt.xticks(np.arange(-1, 3))
plt.yticks(np.arange(-1, 3))
plt.show()
```
