```python
# 연습문제 7.1.1
from sklearn.datasets import load_iris
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
sy = pd.Series(iris.target, dtype='category')
sy = sy.cat.rename_categories(iris.target_names)
df['species'] = sy
df.tail()
```

```python
# 1) 다봉이다
x = df['petal length (cm)'].values
sns.distplot(x)
```

```python
# 2) 단봉이다
x_setosa = df[df.species == "setosa"]["petal length (cm)"].values
sns.distplot(x_setosa)
```

```python
# 연습문제 7.1.2
# 구간이 10개 일때
bins = np.linspace(0,10,11)
x_virginica = df[df.species == 'virginica']['petal length (cm)'].values
x_virginica_mean = np.mean(x_virginica)
x_virginica_median = np.median(x_virginica)
ns, _ = np.histogram(x_virginica, bins = bins)
x_virginica_max = np.argmax(ns)
x_virginica_max_mode = 0.5* (bins[x_virginica_max] + bins[x_virginica_max+1])
print('표본평균', x_virginica_mean)
print('표본중앙값', x_virginica_median)
print('표본최빈값', x_virginica_max_mode)
sns.distplot(x_virginica, bins)
plt.axvline(x_virginica_mean, c = 'k', ls =':', label = '표본평균')
plt.axvline(x_virginica_median, c = 'k', ls ='--', label = '표본중앙값')
plt.axvline(x_virginica_max_mode, c = 'k', ls ='-', label = '표본 최빈값')
plt.show()
```

```python
# 파이썬을 이용한 표본분산 및 표본표준편차의 계산
sp.random.seed(0)
x = sp.stats.norm(0, 2).rvs(1000)  # 평균 =0, 표준편차=2인 정규분포 데이터 생성
print('편향 표본평균 :', np.var(x))
print('편향 표준편차 :', np.std(x))
print('비편향 표본평균 :', np.var(x, ddof=1)) # ddof 인수로 비편향도 구할수 있음
print('비편향 표준편차 :', np.std(x, ddof=1))
```

```python
# 연습문제 7.1.3
import pandas_datareader.data as web

df = web.DataReader('DJCA', data_source='fred').dropna()['DJCA']
r = np.log(df / df.shift(1)).dropna().values
print(r.min())
print(r.max())
```

```python
# 1)
bins = np.linspace(-0.09,0.09,101)
x_mean = np.mean(r)
x_median = np.median(r)
ns, _ = np.histogram(r, bins = bins)
mode_index = np.argmax(ns)
x_mode = 0.5* (bins[mode_index] + bins[mode_index+1])
print('표본평균', x_mean)
print('표본중앙값', x_median)
print('표본최빈값', x_mode)
sns.distplot(r, bins)
plt.axvline(x_mean, c = 'k', ls =':', label = '표본평균')
plt.axvline(x_median, c = 'k', ls ='--', label = '표본중앙값')
plt.axvline(x_mode, c = 'k', ls ='-', label = '표본 최빈값')
plt.show()
```

```python
# 표본첨도
sp.random.seed(0)
x = sp.stats.norm(0, 2).rvs(1000)  # 평균 =0, 표준편차=2인 정규분포 데이터 생성
print('표본비대칭도',sp.stats.skew(x))
print('표본첨도', sp.stats.kurtosis(x))
```

```python
# 표본모멘토
print('1차 모멘토', sp.stats.moment(x, 1)) # 평균
print('2차 모멘토', sp.stats.moment(x, 2)) # 분산
print('3차 모멘토', sp.stats.moment(x, 3)) # 비대칭도
print('4차 모멘토', sp.stats.moment(x, 4)) # 첨도
```
