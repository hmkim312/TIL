```python
# 8.1 scipy를 이용한 확률분포 분석
import scipy as sp
import scipy.stats
```

```python
# 확률분포 클래스
# sp.stats.bernoulli 베르누이 분포
# sp.stats.binom 이항분포
# sp.stats.multinomial 다항분포
# sp.stats.uniform 균일분포
# sp.stats.norm 정규분포
# sp.stats.beta 베타분포
# sp.stats.gamma 감마분포
# sp.stats.t 스튜던트 t분포
# sp.stats.chi2 카이제곱분포
# sp.stats.f F분포
# sp.stats.dirichlet 디리클리분포
# sp.stats.multivariate_normal 다변수 정규분포
```

```python
# 정규분포확률분포를 생성
rv = sp.stats.norm()
# 모수지정
# 기댓값이 1, 표준편차가 2인 정규분포 객체 생성
rv = sp.stats.norm(loc=1, scale=2)
```

```python
# 확률분포 메서드
# rv.pmf 확률질량함수
# rv.pdf 확률밀도함수
# rv.cdf 누적분포함수
# rv.ppf 누적분포함수의 역함수
# rv.sf 생존함수
# rv.isf 생존함수의 역함수
# rv.rvs 랜덤 표본 생성
```

```python
# 확률밀도함수
xx = np.linspace(-8, 8, 100)
pdf = rv.pdf(xx)
plt.plot(xx, pdf)
plt.title('확률밀도함수')
plt.xlabel('$x$')
plt.ylabel('$p(x)$')
plt.show()
```

```python
# 연습문제 8.1.1
rv = sp.stats.norm(loc=0, scale=0.1)
xx = np.linspace(-0.5, 0.5, 100)
pdf = rv.pdf(xx)
plt.plot(xx, pdf)
plt.title('확률밀도함수')
plt.xlabel('$x$')
plt.ylabel('$p(x)$')
plt.show()
```

```python
# 누적분포함수
rv = sp.stats.norm(loc=1, scale=2)
xx = np.linspace(-8, 8, 100)
cdf = rv.cdf(xx)
plt.plot(xx, cdf)
plt.title('누적분포함수')
plt.xlabel('$x$')
plt.ylabel('$F(x)$')
plt.show()
```

```python
# 8.2 베르누이분포와 이항분포
```

```python
# 베르누이 시행 - 결과가 두가지 중 한가지만 나오는 것
# 베르누이 확률변수 - 베르누이 시행을 0또는 1로 바꾼것
# 베르누이 확률분포 - 1이나올 확률인 모수 mu
```

```python
# 사이파이를 사용한 베르누이 확률변수의 시뮬레이션
mu = 0.6
rv = sp.stats.bernoulli(mu)
type(rv)
```

```python
# 확률질량함수
xx = [0, 1]
plt.bar(xx, rv.pmf(xx))
plt.xlim(-1, 2)
plt.ylim(0, 1)
plt.xticks([0, 1], ['x = 0', 'x = 1'])
plt.xlabel('표본값')
plt.ylabel('$P(x)$')
plt.title('베르누이 확률변수의 확률질량함수')
plt.show()
```

```python
# 시뮬레이션 = 표본 생성 작업
x = rv.rvs(100, random_state=0)  # 데이터를 100개 만들어라
x
```

```python
# x를 sns.countplot()으로 시각화
sns.countplot(x)
plt.title('베르누이 확률분포의 시뮬레이션 결과')
plt.xlabel('표본값')
plt.show()
```

```python
# 이론적인 확률분포와 표본의 확률분포를 같이 보기
y = np.bincount(x, minlength=2) / float(len(x))  # 시뮬레이션 값을 확인
df = pd.DataFrame({'이론': rv.pmf(xx), '시뮬레이션': y})  # 데이터프레임으로 생성
df
```

```python
df2 = df.stack().reset_index()  # stack 명령어로 열을인덱스를 나눔
df2.columns = ['표본값', '유형', '비율']  # 열 이름 다시 변경
df2
```

```python
plt.figure(figsize=(7, 5))
sns.barplot(x='표본값', y='비율', hue='유형', data=df2)
plt.title('베르누이분포의 이론적 분포과 시뮬레이션 분포')
plt.show()
```

```python
# 연습문제 8.2.2
def bar_bernoulli(mu, N):
    xx = [0, 1]
    rv = sp.stats.bernoulli(mu)
    x = rv.rvs(N, random_state=10)
    y = np.bincount(x, minlength=2) / float(len(x))
    df = pd.DataFrame({'이론': rv.pmf(xx), f'{N}개의 시뮬레이션 :': y})
    df2 = df.stack().reset_index()
    df2.columns = ['표본값', '유형', '비율']
    plt.figure(figsize=(7, 5))
    sns.barplot(x='표본값', y='비율', hue='유형', data=df2)
    plt.title('베르누이분포의 이론적 분포과 시뮬레이션 분포')
    plt.show()
```

```python
# 1) 확률 0.5, 표본 10개
bar_bernoulli(0.5, 10)
```

```python
# 1) 확률 0.5, 표본 1000개
bar_bernoulli(0.5, 1000)
```

```python
# 2) 확률 0.9, 표본 10개
bar_bernoulli(0.9, 10)
```

```python
# 2) 확률 0.9, 표본 1000개
bar_bernoulli(0.9, 1000)
```

```python
# scipy를 사용한 이항분포의 시뮬레이션
# 그래프 x축의 숫자들은 N * mu
N = 10  # 전체 시행 횟수
mu = 0.6  # 기댓값 모수
rv = sp.stats.binom(N, mu)
xx = np.arange(N+1)
plt.bar(xx, rv.pmf(xx), align=('center'))
plt.xlabel('표본값')
plt.ylabel('$P(x)$')
plt.title('이항분포의 확률질량함수')
plt.show()
```

```python
np.random.seed(0)
x = rv.rvs(100)  # 10번 시행한것을 100번으로 만듬, 6은 1이 6번 나왔다는 의미
x
```

```python
sns.countplot(x)
plt.title('이항분포의 시뮬레이션 결과')
plt.xlabel('표본값')
plt.show()
```

```python
y = np.bincount(x, minlength=N+1) / len(x)
df = pd.DataFrame({'이론': rv.pmf(xx), '시뮬레이션': y}).stack().reset_index()
df.columns = ['표본값', '유형', '비율']
df
```

```python
sns.barplot('표본값', '비율', '유형', data=df)
plt.title('이항분포의 이론적분포와 시뮬레이션 분포')
plt.show()
```

```python
# 연습문제 8.2.3
def bar_binom(mu, N, NN):
    xx = np.arange(N+1)
    N = N  # 전체시행 횟수
    mu = mu  # 기댓값 모수
    NN = NN  # 표본의 갯수
    rv = sp.stats.binom(N, mu)
    np.random.seed(0)
    x = rv.rvs(NN)  # N번 시행한것을 NN개의 표본으로 만듬
    y = np.bincount(x, minlength=N+1) / len(x)
    df = pd.DataFrame(
        {'이론': rv.pmf(xx), f'{N}번시행,{NN}개의 표본시뮬레이션': y}).stack().reset_index()
    df.columns = ['표본값', '유형', '비율']
    sns.barplot('표본값', '비율', '유형', data=df)
    plt.title(f'이항분포의 이론적분포와 {N}번시행,{NN}개의 표본시뮬레이션 분포')
    plt.show()
```

```python
# 1) 확률 = 0.5, 시행횟수 5, 표본갯수 10
bar_binom(0.5, 5, 10)
```

```python
# 1) 확률 = 0.5, 시행횟수 5, 표본갯수 1000
bar_binom(0.5, 5, 1000)
```

```python
# 2) 확률, 0.9, 시행횟수 20, 표본갯수 10
bar_binom(0.9, 20, 10)
```

```python
# 2) 확률 = 0.9, 시행횟수 20, 표본갯수 1000
bar_binom(0.9, 20, 1000)
```
